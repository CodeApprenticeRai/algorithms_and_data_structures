```python
#The ordered list IMO subclasses the inordered list,
#with the except that the insert method uses some ordering constraint
#to update the list

#Node Object
class OrderedListNode:
    def __init__(self, initdata):
        self.data = initdata
        self.next = None
        
    def setData(self, data):
        self.data = data
        
    def getData(self):
        return self.data
    
    def setNext(self,next_node):
        self.next = next_node
    
    def getNext(self):
        return self.next

#
class OrderedList:
    def __init__(self):
        self.head = None
    
    def add(self,data):
        current = self.head
        new_node = OrderedListNode(data)
        
        while (True):
            #case 0: the list is empty: create new node and have it
            # link to the empty tail, then make it the head of the
            # list
            if current == None: 
                new_node.setNext(self.head)
                self.head = new_node
                return True
            
            #case 1: addition is at head: link the new_node to head
            #and then make new_node the head
            if (current.getData() > data) and (current == self.head):
                new_node.setNext(self.head)
                self.head = new_node
                return True
            
            #case 2: Duplicate 
            if current.getData() == data:
                return "Duplicate"
            
            #case 3: the list isn't empty: search one node ahead
            # for a number greater than the number being inserted
            # then insert directly behind that number.
            # Get greater_number_node -> link current_node to 
            # node_being_inserted -> linnk node being inserted
            #to current node.
            if ( current.getNext().getData() > data ):
                in_front_of_new_node = current.getNext()
                current.setNext(new_node)
                new_node.setNext(in_front_of_new_node)
                return True
            
            current = current.getNext()  
        
    def search(self, item):
        current = self.head
        
        while current.data != item: 
            if current.data == None:
                return False
            
            if current.data == item:
                return True
            
            current = current.getNext()

    def remove(self, item):
        current = self.head
        found = False
        
        while not found:
            
            if current == None:
                return "Item Not Present in List"
            
            #Search one node ahead for the node we want to remove
            #n1-->n2...-->current--->node_being_removed--->next_node
            if current.getPrev().getData() == item:
                node_being_removed = current.getPrev()
                
                found = True 
                
                #overwrite the node being removed with the node ahead of it
                #n1-->n2...-->current--->next_node
                current.setPrev( node_being_removed.getPrev() ) 
                return True
            
            current = current.getPrev()
            
    def printList(self, count=0):
        current = self.head
        
        
        while current != None:
            print(current.getData(), "|", sep="\n")
            count += 1
            current = current.next
            
        return
    
```

```python
import random as r

r.seed(0)

uol = OrderedList()

for i in range(10):
    uol.add(int( r.random()*100) )
    
uol.printList()
```
